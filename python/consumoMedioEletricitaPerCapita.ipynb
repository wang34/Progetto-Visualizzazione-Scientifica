```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

dati = pd.read_csv('..\csv\consumi-e-produzioni\per-capita-energy-use\per-capita-energy-use.csv')
df = pd.DataFrame(dati)

df=df[(df['Year']>1979) & (df['Year']<2023)] 

print(df)

df=df[(df['Entity']=="United States") | (df['Entity']=="Iceland")
      | (df['Entity']=="Norway")
      | (df['Entity']=="Europe")
      | (df['Entity']=="China")
      | (df['Entity']=="Canada")
      | (df['Entity']=="Italy")
      | (df['Entity']=="India")
      ]
#print(df.groupby("Entity"))

x = np.arange(1980, 2023, 1)
#print(x.size)

y = df[(df['Entity']=="Iceland")]
y= y['Primary energy consumption per capita (kWh/person)']
#print(y.size)

y1= df[(df['Entity']=="Norway")]
y1= y1['Primary energy consumption per capita (kWh/person)']

y2= df[(df['Entity']=="Canada")]
y2= y2['Primary energy consumption per capita (kWh/person)']

y3= df[(df['Entity']=="United States")]
y3= y3['Primary energy consumption per capita (kWh/person)']

y4= df[(df['Entity']=="China")]
y4= y4['Primary energy consumption per capita (kWh/person)']

y5= df[(df['Entity']=="Europe")]
y5= y5['Primary energy consumption per capita (kWh/person)']

y6= df[(df['Entity']=="Italy")]
y6= y6['Primary energy consumption per capita (kWh/person)']

y7= df[(df['Entity']=="India")]
y7= y7['Primary energy consumption per capita (kWh/person)']
# **Aggiungi questa linea per cambiare le dimensioni del grafico**
plt.figure(figsize=(12, 6))  # Larghezza 12 pollici, Altezza 6 pollici. Regola questi valori a piacere.

plt.grid(True, which='both', axis='both', linestyle='--', color='gray', alpha=0.5)
plt.xticks(np.arange(1980, 2024, 5)) 
#plt.xticks(np.arange(1800, 2023, 20)) 

# Creazione del grafico
plt.plot(x, y, label='Iceland', color='darkblue',linewidth=2)
plt.plot(x, y3, label='United States',color='red',linewidth=2)
plt.plot(x, y4, label='China',color="gold",linewidth=2)
plt.plot(x, y5, label='Europe',color="skyblue",linewidth=2)
plt.plot(x, y6, label='Italy',color="forestgreen",linewidth=2)
plt.plot(x, y7, label='India',color="brown",linewidth=2)

#plt.plot(x, y1, label='Altri',color="black")  
#plt.plot(x, y2, color="black")  
#plt.plot(x, y5, color="black")  
#plt.plot(x, y6,color="black")  
#plt.plot(x, y7,color="black")  

#plt.title('consumo medio di elettricit per capita :1980 --> 2024')

# Etichette per gli assi
plt.ylabel('Kwh')

# Aggiungi la legenda
plt.legend()

# Mostra il grafico
plt.show()
```
