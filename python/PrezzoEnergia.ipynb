```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

# Caricare i dati
dati = pd.read_csv('..\csv\prezzo\estat_ten00117_en.csv',usecols = [4,5,6,7,8,9])
df = pd.DataFrame(dati)

paesi_interesse = [
    "Italy",
    "European Union - 27 countries (from 2020)",
    "Spain",
    "Germany",
    "France",
    "Poland"
]
df = df[df["geo"].isin(paesi_interesse)]
df = df[df["indic_en"]=="Medium size households"] #medium size consumers
# **Aggiungi questa linea per cambiare le dimensioni del grafico**
plt.figure(figsize=(12, 6))  # Larghezza 12 pollici, Altezza 6 pollici. Regola questi valori a piacere.
print(df)

# Dati da visualizzare
x = np.arange(2013,2025,1)


y1= df[df["geo"]== "Italy"]
y1= y1['OBS_VALUE']


y2= df[df["geo"]== "European Union - 27 countries (from 2020)"]
y2= y2['OBS_VALUE']


y3= df[df["geo"]== "Spain"]
y3= y3['OBS_VALUE']


y4= df[df["geo"]== "Germany"]
y4= y4['OBS_VALUE']


y5= df[df["geo"]== "France"]
y5= y5['OBS_VALUE']


y6= df[df["geo"]== "Poland"]
y6= y6['OBS_VALUE']


plt.grid(True, which='both', axis='both', linestyle='--', color='gray', alpha=0.5)
plt.xticks(np.arange(2000, 2025, 1))

# Creazione del grafico

plt.plot(x, y1, label='Italy',color='forestgreen',linewidth=2)  # Prima linea
plt.plot(x, y2, label='European Union - 27 countries (from 2020)',color='skyblue',linewidth=2)  # Seconda linea
plt.plot(x, y3, label='Spain',color='brown',linewidth=2 )  # Terza linea
plt.plot(x, y4, label='Germany',color='gold', linewidth=2)
plt.plot(x, y5, label='France',color='blue',linewidth=2 )
plt.plot(x, y6, label='Poland',color='red',linewidth=2)
# Aggiungi un titolo
plt.title('Prezzo dell\'elettricit per Kilowatt-hour')

# Etichette per gli assi
plt.ylabel('euro')

# Aggiungi la legenda
# **Modifica principale: Sposta la legenda fuori dal grafico a destra**
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Mostra il grafico
plt.show()
```
